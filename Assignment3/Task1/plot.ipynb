# *Plotty*

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(x1, y1, x2, y2, title, x, y):
    plt.plot(x1, y1, label="Strategy 1")
    plt.plot(x2, y2, label="Strategy 2")
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()
    save_title = title.replace(" ","_").lower()
    plt.savefig(f"img/{save_title}.png")
    plt.show()
```

### Throughput

```python
data_s1_thr = np.loadtxt("output/s1_thr.out")
x1 = data_s1_thr[:, 0]
y1 = data_s1_thr[:, 1]

data_s2_thr = np.loadtxt("output/s2_thr.out")
x2 = data_s2_thr[:, 0]
y2 = data_s2_thr[:, 1]


plot(x1, y1, x2, y2, "Throughput vs Network Load", "Network Load", "Throughput")
```

### Lossrate

```python
data_s1_loss = np.loadtxt("output/s1_loss.out")
x1 = data_s1_loss[:, 0]
y1 = data_s1_loss[:, 1]

data_s2_loss = np.loadtxt("output/s2_loss.out")
x2 = data_s2_loss[:, 0]
y2 = data_s2_loss[:, 1]


plot(x1, y1, x2, y2, "Loss rate vs Network Load", "Network Load", "Loss rate")
```

### Varying radius

```python
data_radius = np.loadtxt("output/radius.out")
x1 = data_radius[:, 0]
y1 = data_radius[:, 1]

title = "Throughput vs Radius"
plt.plot(x1, y1)
plt.title(title)
plt.xlabel("Radius")
plt.ylabel("Throughput")
save_title = title.replace(" ", "_").lower()
plt.savefig(f"img/{save_title}.png")
plt.show()
```
